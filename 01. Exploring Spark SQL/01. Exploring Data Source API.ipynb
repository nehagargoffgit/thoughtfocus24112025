#Csutome Schema

```python
from pyspark.sql.types import *
 
USER_SCHEMA = StructType(
    [
        StructField("id", IntegerType()),
        StructField("age", IntegerType()),
        StructField("gen", StringType()),
        StructField("designation", StringType()),
        StructField("salary", IntegerType()),
    ]
)
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    schema=USER_SCHEMA
)
df.printSchema()
df.display()
```

####Handling Bad record like in json value is missing 
1. PERMISSIVE (DEFAULT) - BAD RECORD WILL GO IN CORRUP_RECORD 
2. DROPMALFORMED - DROP RECORD
3. FAILFAST - NOT ALLOW TO CREATE FRAME

####PERMISSIVE

```python
spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/access_logs.json",
    mode="PERMISSIVE"
).display()
```

#Write DataFrame in File Default errorIfExist

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
print(type(df.write))
# Default mode is error , so on again try for same file to save it give error 
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
)
spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json"
).show()
```

#Overwrite Mode

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
print(type(df.write))
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
    mode='overwrite'
)
 
 
spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json"
).show()
 
```

#Append mode create a new file or add it

```python
df = spark.read.csv(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/sample_dataset/users_001.csv",
    header=True,
    inferSchema=True,
)
print(type(df.write))
df.write.save(
    format="JSON",
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json",
    mode='append'
)
 
 
spark.read.json(
    path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_json"
).show()
```
