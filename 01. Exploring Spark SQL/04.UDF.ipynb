##Prerequisite

##Load employee.csv in DataFrame

```python
df = spark.read.csv(path="/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/employee.csv",
                    sep="|",
                    inferSchema=True,
                    header=True,
                    quote="'")
df.display()
```

```python
def conver_to_title(st):
  if st is None:
    return None
  else:
    return st.title()
    
```

```python
df.select("id",conver_to_title("name")).display()
```

#USE Python Function

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
convert_to_title_udf = udf(conver_to_title, StringType())
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType
convert_to_title_udf = udf(lambda e:conver_to_title(e) if e is not None else None, StringType())
```

```python
df.select("id",convert_to_title_udf("name")).display()
```

```python
df.createOrReplaceTempView("employee_vw")
```

####to use udf in sql we need to register udf to use inside to sql

```python
spark.udf.register("convert_to_title_udf", conver_to_title,StringType())
#spark.sql("select id, convert_to_title(name) from employee_vw").display()
```
