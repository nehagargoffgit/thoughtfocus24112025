#Prerequisite

```python
df =spark.read.csv(path='/Volumes/quickstart_catalog/quickstart_schema/sandbox/dataset/user_dataset/users_001.csv',
                   header=True,
                   inferSchema=True)
df.display()
```

```python
df.write.format("parquet").save('/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet',mode='overwrite')

df_parquet =spark.read.parquet(
    "/Volumes/quickstart_catalog/quickstart_schema/sandbox/output_parquet"
)
df_parquet.display()
```

#Create View

```python
df_parquet.createOrReplaceTempView("user_vw")
```

#Disadvantage - Update not allowed

```python
%sql

Update user_vw
Set country ='Ind'
where country ='India'
```
