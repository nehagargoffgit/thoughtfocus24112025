#Bronze Layer
medallion_staging --> Bronze Table

###Customers

```python
spark.read.csv(
    "/Volumes/dev_catalog/default/retail_staging/customers/customer.csv",
    header=True,
    inferSchema=True,
).write.format("delta").mode("overwrite").saveAsTable("dev_catalog.bronze.customers")
```

```python
spark.read.table('dev_catalog.bronze.customers').display()
#spark.read.table('dev_catalog.bronze.customers')
```

###Orders

```python
spark.read.csv(
    "/Volumes/dev_catalog/default/retail_staging/orders/",
    header=True,
    inferSchema=True,
).write.format("delta").mode("overwrite").saveAsTable("dev_catalog.bronze.orders")
```

```python
spark.read.table('dev_catalog.bronze.orders').display()
```

#Silver Layer

```python
from pyspark.sql.functions import col
spark.read.table('dev_catalog.bronze.customers').filter(col('CustomerName').isNotNull()).write.mode("append").saveAsTable('dev_catalog.silver.customers')
spark.read.table('dev_catalog.silver.customers').display()
```

###Orders

```python
from pyspark.sql.functions import col
spark.read.table('dev_catalog.bronze.orders').filter(col('CustomerID').isNotNull()).write.mode("append").saveAsTable('dev_catalog.silver.orders')
spark.read.table('dev_catalog.silver.orders').display()
```

#Gold Layer

```python
customers_df = spark.read.table('dev_catalog.silver.customers')
orders_df = spark.read.table('dev_catalog.silver.orders')
join_df = customers_df.join(orders_df, customers_df.CustomerID == orders_df.CustomerID,how='inner')
join_df.display()
join_df.select(customers_df.CustomerID,"orderID").groupBy("CustomerID").count().write.mode("append").saveAsTable('dev_catalog.gold.sales_agg')
spark.read.table('dev_catalog.gold.sales_agg').display()
```
