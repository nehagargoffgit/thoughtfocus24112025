#Gold Layer

```python
customers_df = spark.read.table('dev_catalog.silver.customers')
orders_df = spark.read.table('dev_catalog.silver.orders')
join_df = customers_df.join(orders_df, customers_df.CustomerID == orders_df.CustomerID,how='inner')
join_df.display()
join_df.select(customers_df.CustomerID,"orderID").groupBy("CustomerID").count().write.mode("append").saveAsTable('dev_catalog.gold.sales_agg')
spark.read.table('dev_catalog.gold.sales_agg').display()
```
